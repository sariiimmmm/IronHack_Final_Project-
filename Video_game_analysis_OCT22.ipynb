```python
mac = []
windows = []
linux = []
for i in vg4['platforms']:
    if 'Mac' in i:
        mac.append('Mac')
    else:
        mac.append("")
    if 'Windows' in i:
        windows.append('Windows')
    else:
        windows.append("")
    if 'Linux' in i:
        linux.append('Linux')
    else:
        linux.append("")
```

```python
vg4['platform1'] = mac
vg4['platform2'] = windows
vg4['platform3'] = linux
```
